```python
def geracao_texto(mensagens, model="llama-3.1-8b-instant", max_tokens=1000, temperature=0):
    resposta= client.chat.completions.create(
        messages=mensagens,
        model=model,
        max_tokens=max_tokens,
        temperature=temperature
    )
    mensagens.append({"role":"assistant","content":resposta.choices[0].message.content})
    return mensagens
```

```python
msg1=[{"role":"user","content":'defina o que é api em 5 palavras:'}]
mensagens= geracao_texto(msg1)
```

### Utilizando Stream

```python
mensagens = [{"role": "user", "content": "conte uma história simples e objetiva com LLM"}]

resposta = client.chat.completions.create(
    messages=mensagens,
    model="llama-3.1-8b-instant",
    max_tokens=1000,
    temperature=0,
    stream=True
)

for resposta_strem in resposta:
    texto = resposta_strem.choices[0].delta.content  #O .delta (delta = variação/mudança) existe justamente porque é stream — ele carrega só o trecho novo, não a resposta inteira.

    if texto:
        print(texto, end="")
```
